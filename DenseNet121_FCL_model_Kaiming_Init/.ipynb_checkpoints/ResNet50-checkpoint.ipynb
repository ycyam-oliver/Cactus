```python
import os
from PIL import Image
import torch as t
import torchnet.meter as meter
import torch.nn as nn
import torch.nn.functional as F
from torch import optim
from torch.utils import data
from torch.utils.data import Dataset, DataLoader
from torch.autograd import Variable
import numpy as np
import torchvision as tv
from torchvision import transforms as T
from torchvision.transforms import ToPILImage
from torchvision import utils
from torchvision import models
import pandas as pd
from pandas import Series,DataFrame
import visdom
import matplotlib.pyplot as plt
```

# Import Data

```python
class Cactus(data.Dataset):
    
    def __init__(self,root,transforms=None, train=True, test=False):
        
        self.test=test
        
        if root=='test':
            df=pd.read_csv('sample_submission.csv')
        else:
            df=pd.read_csv(root+'.csv')

        imgs=(df['id']).values
        imgs=[os.path.join(root, img) for img in imgs]
        imgs_num=len(imgs)
        
        labels=(df['has_cactus']).values
        
        if self.test:
            self.imgs=imgs
            self.labels=(df['id']).values
                
        elif train:
            self.imgs=imgs[:int(0.7*imgs_num)]
            self.labels=labels[:int(0.7*imgs_num)]
            # store some weights info for Sampler
            count_0=sum(self.labels==0)
            count_1=sum(self.labels==1)
            # inverse the number of sample in each class
            # to be the weight of sampling
            self.weight_of_labels=[
                1/count_1 if cac==1 else 1/count_0 for cac in self.labels]
            
        
        else: #validation set
            self.imgs=imgs[int(0.7*imgs_num):]
            self.labels=labels[int(0.7*imgs_num):]
        
        if transforms is None:
            
            normalize=T.Normalize(
                mean=[0.485, 0.456, 0.406],
                std=[0.229, 0.224, 0.225])

            if self.test or not train:
                
                # make the size fit the transfer model
                
                self.transforms=T.Compose([
                    T.Resize(224),
                    T.ToTensor(),
                    normalize])
                
            else: # if train
                
                # some data augmentations
                # and make the size fit the transfer model
                
                self.transforms=T.Compose([
                    T.RandomVerticalFlip(),
                    T.RandomHorizontalFlip(),
                    T.ColorJitter(),
                    T.Resize(224),
                    T.ToTensor(),
                    normalize])
            
    def __getitem__(self, index):
        
        img_path=self.imgs[index]
        
        label=self.labels[index]
        
        data=Image.open(img_path)
        data=self.transforms(data)
        
        return data, label
    
    def __len__(self):
        return len(self.imgs)
```

```python
# load the data into train and test sets

batchsize=32

train_dataset=Cactus('train',train=True)
# Use WeightedRandomSampler to make the training set more balanced
weights=t.FloatTensor(train_dataset.weight_of_labels)
train_sampler=t.utils.data.WeightedRandomSampler(
    weights,len(train_dataset),replacement=True)
trainloader=DataLoader(train_dataset,batch_size=batchsize,
                       sampler=train_sampler,
                       shuffle=False)

val_dataset=Cactus('train',train=False)
valloader=DataLoader(val_dataset,batch_size=batchsize,shuffle=True)

test_dataset=Cactus('test',train=False,test=True)
testloader=DataLoader(test_dataset,batch_size=batchsize,shuffle=False)
```

```python
# a function to undo the normalization for showing purpose

class UnNormalize(object):
    def __init__(self, mean, std):
        self.mean = mean
        self.std = std

    def __call__(self, tensor):
        """
        Args:
            tensor (Tensor): Tensor image of size (C, H, W) to be normalized.
        Returns:
            Tensor: Normalized image.
        """
        for t, m, s in zip(tensor, self.mean, self.std):
            t.mul_(s).add_(m)
            # The normalize code -> t.sub_(m).div_(s)
        return tensor

unorm=UnNormalize(
    mean=(0.485, 0.456, 0.406), std=(0.229, 0.224, 0.225))    
    
```

```python
# see what the images look like

dataiter=iter(trainloader)
images,labels=dataiter.next() 
# will give out 'batch_size' number of images 

for im in range(len(images)):
    
    # print the label
    print('Label=',str(labels[im].item()),':')
    
    plt.figure()
    
    image=(unorm(images[im])).permute(1,2,0)
    # the format of (data+1)/2 is just to undo the normalization
    # for showing purpose
    plt.imshow(image)
    plt.show()
```

# Model

```python
model=models.resnet50(pretrained=True)

# gonna train only last few layers
# so freeze the model weights
for param in model.parameters():
    param.requires_grad=False

# use our self-defined classifier as the head

num_filters=model.fc.in_features

model.fc=nn.Sequential(nn.Linear(num_filters, 256),
                                  nn.ReLU(),
                                  nn.Dropout(0.4),
                                  nn.Linear(256,2),
                                  nn.Sigmoid())

model.fc
```

# Training

```python
# define an optimizer

criterion = nn.CrossEntropyLoss()
optimizer=optim.SGD(model.parameters(), lr=0.001, momentum=0.9)

loss_meter=meter.AverageValueMeter()
confusion_matrix=meter.ConfusionMeter(2)
previous_loss=1e100
```

```python
# define a function of validation for the use in training
def val(model,dataloader):
    
    # validation mode
    model.eval()
    
    confusion_matrix=meter.ConfusionMeter(2)
    for ii,data in enumerate(dataloader):
        
        # input info:
        
        inputs, labels=data
        # inputs has batchsize number of images w 3 channels
        
        val_inputs=Variable(inputs)
        val_labels=Variable(labels)
        # Variable has also grad info
        
        outputs=model(val_inputs)
        confusion_matrix.add(outputs.data.squeeze(),labels)
        
    # resume the training mode for the model
    model.train()
    
    cm_value=confusion_matrix.value()
    
    TN=cm_value[0][0]
    FP=cm_value[0][1]
    FN=cm_value[1][0]
    
    precision=TN/(TN+FN)
    recall=TN/(TN+FP)
    
    # F1 score for label '0'('negative')
    F1_score=2*precision*recall/(precision+recall)
    
    return confusion_matrix, F1_score
```

```python
# training

max_epoch=10

val_F1_plot=[]
val_loss_plot=[]

for epoch in range(max_epoch):
    
    loss_meter.reset()
    confusion_matrix.reset()
    
    for ii,(data,label) in enumerate(trainloader):
        
        inputs=Variable(data)
        # inputs has bathsize number of images w 3 channels
        # note Variable has also grad info
        target=Variable(label)
        
        # clear the gradient
        optimizer.zero_grad()
        
        # forward & backward
        outputs=model(inputs)
        loss=criterion(outputs, target)
        loss.backward()
        
        # update the parameters
        optimizer.step()
        
        # gather statistic score and visualize things
        loss_meter.add(loss.data.item())
        confusion_matrix.add(outputs.data, target.data)
        
        print_every_batch=100
        if ii%print_every_batch==(print_every_batch-1):
            # print the training process every print_every_batch batch
            print('[epoch=%d, data index=%5d] loss: %1.3f' %(
                epoch+1,ii+1,loss_meter.value()[0]))
    
    val_cm,val_F1=val(model,valloader)
    print('Validation F1 score is ',val_F1)
    val_F1_plot+=[val_F1]
    
    # lower lr if loss is not getting down
    lr_decay=0.95
    lr=0.001
    if loss_meter.value()[0]>previous_loss:
        lr=lr*lr_decay
        for param_group in optimizer.param_groups:
            param_group['lr']=lr
            
    previous_loss=loss_meter.value()[0]
    val_loss_plot+=[previous_loss]
```

```python
# see how the validation accuracy and loss propogate w epoch
plt.plot(val_F1_plot,label='Validation F1 score')
plt.plot(val_loss_plot,label='Validation Loss')
plt.legend()
plt.xlabel('epoch')
plt.show()

print('The final F1 score is %1.5f'%val_F1_plot[-1])
print('The final Loss is %1.5f'%val_loss_plot[-1])
```

# Test Set

```python
pic_id=[]
results=[]
for ii,(data,path) in enumerate(testloader):
    inputs=t.autograd.Variable(data)
    outputs=model(inputs)
    probability=t.sigmoid(outputs)[:,1].data.tolist()
    
    batch_id=list(path)
    pic_id+=batch_id
    
    batch_results=[1 if prob_i>0.5 else 0 for prob_i in probability]
    results+=batch_results
```

```python
# save the predicted results to a csv file

submit=pd.DataFrame({"id":pic_id,"has_cactus":results})
submit.to_csv('predicted_labels_ResNet50.csv',index=False)
```

```python
# check some pics using the trained network
dataiter=iter(testloader)

images, path=dataiter.next()

# print the predicted labels using the network
image_labels=[list(
    submit.loc[submit['id']==i,'has_cactus'])[0] for i in path]

for im in range(len(images)):
    
    # print the label
    print('Predicted label=',str(image_labels[im]),':')
    
    plt.figure()
    
    image=(unorm(images[im])).permute(1,2,0)
    # the format of (data+1)/2 is just to undo the normalization
    # for showing purpose
    plt.imshow(image)
    plt.show()
```
